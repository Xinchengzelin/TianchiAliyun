# Model

```python
# scaler = MinMaxScaler(feature_range=(0,1))
# train_select=train.loc[:,[col for col in train.columns if col != 'target']]
# train_scale = pd.DataFrame(scaler.fit_transform(train_select), columns=train_select.columns)
# test_scale = pd.DataFrame(scaler.transform(test), columns=test.columns)
# X_train,X_test,y_train,y_test=train_test_split(train_scale,train['target'])

X_train=train.loc[:,[col for col in train.columns if col != 'target']]
y_train=train['target']
scaler = MinMaxScaler(feature_range=(0,1))
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(test), columns=test.columns)
```

```python
X_train.shape,X_test.shape,y_train.shape
```

```python
X_train.head()
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
mean_squared_error(y_train,model.predict(X_train))

```

```python
#训练集分出部分测试集
# model_ridge=Ridge(alpha=4,random_state=111)
# model_ridge.fit(X_train,y_train)
# scores = mean_squared_error(y_test,model_ridge.predict(X_test)) 
# print(scores)
```

```python
model_ridge=Ridge(alpha=250,random_state=111)
model_ridge.fit(X_train,y_train)
scores = cross_val_score(model_ridge, X_train, y_train, cv=5, scoring="neg_mean_squared_error") 
print(scores.mean(),'***',mean_squared_error(y_train,model_ridge.predict(X_train)),scores)

# model_ridge.fit(X_train,y_train)
```
